```python
train = pd.read_csv("train.csv", parse_dates=["datetime"])
train.head()
```

```python
train.describe()
```

```python
test = pd.read_csv("test.csv", parse_dates=["datetime"])
test.head()
```

```python
submission = pd.read_csv("sampleSubmission.csv", parse_dates=["datetime"])
submission.head()
```

```python
train_data = train.drop(columns=["casual", "registered"])
```

```python
predictor = TabularPredictor(
    label="count", 
    eval_metric="root_mean_squared_error"
).fit(
    train_data=train_data, 
    time_limit=600, 
    presets="best_quality"
)
```

```python
predictor.fit_summary()
```

```python
predictions = predictor.predict(test)
predictions.head()
```

```python
predictions.describe()
```

```python
(predictions < 0).sum()
```

```python
submission["count"] = predictions
submission.to_csv("submission.csv", index=False)
```

```python
!kaggle competitions submit -c bike-sharing-demand -f submission.csv -m "first raw submission"
```

```python
train["hour"] = train["datetime"].dt.hour
test["hour"] = test["datetime"].dt.hour
```

```python
train["season"] = train["season"].astype("category")
train["weather"] = train["weather"].astype("category")

test["season"] = test["season"].astype("category")
test["weather"] = test["weather"].astype("category")
```

```python
train.head()
```

```python
train.hist(figsize=(15, 12), bins=30)
```

```python
train_new = train.drop(columns=["casual", "registered"])
```

```python
predictor_new_features = TabularPredictor(
    label="count",
    eval_metric="root_mean_squared_error"
).fit(
    train_data=train_new,
    time_limit=600,
    presets="best_quality"
)
```

```python
predictor_new_features.fit_summary()
```

```python
predictions_new = predictor_new_features.predict(test)
```

```python
predictions_new[predictions_new < 0] = 0
```

```python
submission_new_features = pd.read_csv("sampleSubmission.csv")
submission_new_features["count"] = predictions_new
submission_new_features.to_csv("submission_new_features.csv", index=False)
```

#### New Score of `0.62619`

```python
train_hpo = train.drop(columns=["casual", "registered"])
```

```python
hyperparameters = 'default'
hyperparameter_tune_kwargs = {
    'num_trials': 3,
    'scheduler': 'local',
    'searcher': 'random'
}
predictor_new_hpo = TabularPredictor(
    label="count",
    eval_metric="root_mean_squared_error"
).fit(
    train_data=train_hpo,
    time_limit=600,
    presets="best_quality",
    hyperparameters=hyperparameters,
    hyperparameter_tune_kwargs=hyperparameter_tune_kwargs
)
```

```python
predictor_new_hpo.fit_summary()
```

```python
predictions_hpo = predictor_new_hpo.predict(test)
predictions_hpo[predictions_hpo < 0] = 0
```

```python
submission_new_hpo = pd.read_csv("sampleSubmission.csv")
submission_new_hpo["count"] = predictions_hpo
submission_new_hpo.to_csv("submission_new_hpo.csv", index=False)
```

```python
!kaggle competitions submit -c bike-sharing-demand -f submission_new_hpo.csv -m "new features with hyperparameters"
```
